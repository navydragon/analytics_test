```python
import pandas as pd
df = pd.read_csv('cars_2.csv')
df.info()
```

```python
df2 = df[df['год выпуска'] >= 2014].reset_index()
df2.info()
df2.head()
```

```python
def get_style(text):
    if type(text) == str:
        return str(text[13:-25]).split("', '")
    else:
        print(text)
df2['style'] = df2['name'].apply(get_style)
```

```python
df2['brand'] = df2['style'].apply(lambda x: x[0] if type(x)==list else print(x))
df2['model'] = df2['style'].apply(lambda x: x[1] if type(x)==list else print(x))
df2['generation'] = df2['style'].apply(lambda x: x[2] if type(x)==list else print(x))
df2['style'] = df2['style'].apply(lambda x: x[4] if type(x)==list else print(x))
```

```python
def parse_km(text):
    if type(text) == str:
       res = text.replace("км", "")
       res = res.replace("\xa0", "")
       res = int(res)
       if res < 1000:
           res = 0
       return res 
    else:
        print(text)
df2['Пробег'] = df2['Пробег'].apply(parse_km)
df2.head()
```

```python
df2 = df2.drop(columns=['index', 'name','ПТС','Таможня','Комплектация','numberOfDoors','Владение','Владельцы'])
df2.head()
```

```python
df2.to_excel ('cars.xlsx', index = None, header=True)
```
